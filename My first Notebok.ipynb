```python
import urllib3

link = "https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=1999&sort=desc&view=detailed"

# Construct a python request pool
urllib3.disable_warnings()
http = urllib3.PoolManager(cert_reqs='CERT_NONE')

# Initiate a web request
r = http.request('GET', link, headers={'User-Agent': 'Mozilla/5.0'})

# Convert the result from raw bytes to text
datastring = str(r.data, "utf-8")

# Check whether it succeeded and how much data we got from the site
print(r.status)
print(len(datastring))
```

```python
import re
# Search for titles
title = re.findall("class=\"title\"><h3>(.*)<\/a>", datastring)

#search release dates
Release = re.findall("<div class=\"clamp-details\">\s+ <span>(.*)<\/span>", datastring)


#search descriptions
Description = re.findall("<div class=\"summary\">\n\s+(.*)", datastring)


#search metascores
meta = re.findall("<div class=\"metascore_w large movie positive\">(.*)"
                "<\/div>\n\s+<\/a>\n\s+<\/div>\n\s+<div class=\"clamp-userscore\">", datastring)
     

#search thumbnails
thumb = re.findall("<img src=\"(.*)\" alt=\"(?:.*)\" \/><\/a>", datastring)


#using for loop to append found information to create a movie list. 
movies = []

for i in range(len(title)):
    movies.append("Movie Title: " + title[i] + "\n"
                "Release Date: " + Release[i] + "\n"
                "Summary: " + Description[i] + "\n"
                "Metascore: " + meta[i] + "\n"
                "Image URL: " + thumb[i])
#while loop to append found information to create a movie list.
i = 0
while i <len(movies):
    print(movies[i] + "\n")
    i = i +1                






```
